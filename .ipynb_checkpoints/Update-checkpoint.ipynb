```python
Type_List_01 = pd.Series({'CU':'沪铜',
            'NI':'沪镍',
            'J9':'焦炭',
            'PB':'沪铅',
            'JM':'焦煤',
            'I9':'铁矿',
            'JD':'鸡蛋',
            'RU':'橡胶',
            'Y9':'豆油',
            'SR':'白糖',
            'AL':'沪铝',
            'HC':'热卷',
            'RB':'螺纹',
            'ZC':'郑煤',
            'MA':'甲醇',
            'ZN':'沪锌',
            'RM':'菜粕',
            'C9':'玉米',
            'M9':'豆粕'})
```

```python
for name in Type_List_01.index:
    try:
        DF = pd.read_csv('1hour\\'+name+'1h.csv',sep=',',encoding='gb2312')
        DF = DF.iloc[:,0:6]
        DF.columns = ['time','open','high','low','close','vol'] 
        print(Type_List_01[name]+':'+str(DF.time[0]))
    except:
        print(name+'Wrong')
```
